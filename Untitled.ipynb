```python
import os 
import cv2
from datetime import datetime
import pandas as pd
# import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import nnabla as nn
import nnabla.functions as F
import nnabla.parametric_functions as PF
from nnabla.utils.data_iterator import data_iterator_csv_dataset 

cap = cv2.VideoCapture(0) # 任意のカメラ番号に変更する
#realtimeというフォルダに保存していく。
new_dir_path = "./realtime/"
os.makedirs(new_dir_path, exist_ok=True)


print('始めるには入力して')
C = input()
no=0
def network(x, y, test=False):
    # Input:x -> 1,60,120
    # Convolution -> 10,57,116
    h = PF.convolution(x, 10, (4,5), (0,0), name='Convolution')
    # MaxPooling -> 10,14,23
    h = F.max_pooling(h, (4,5), (4,5))
    # Tanh
    h = F.tanh(h)

    # Convolution_2 -> 8,10,19
    h1 = PF.convolution(h, 8, (5,5), (0,0), name='Convolution_2')

    # Convolution_3 -> 8,10,19
    h2 = PF.convolution(h, 8, (5,5), (0,0), name='Convolution_3')
    # MaxPooling_2 -> 8,5,9
    h1 = F.max_pooling(h1, (2,2), (2,2))
    # MaxPooling_3 -> 8,5,9
    h2 = F.max_pooling(h2, (2,2), (2,2))
    # Tanh_2
    h1 = F.tanh(h1)
    # Tanh_4
    h2 = F.tanh(h2)

    # Mul2 -> 8,5,9
    h1 = F.mul2(h1, h2)
    # Affine -> 10
    h1 = PF.affine(h1, (10,), name='Affine')
    # Tanh_3
    h1 = F.tanh(h1)
    # Affine_2 -> 3
    h1 = PF.affine(h1, (3,), name='Affine_2')
    # Softmax
    h1 = F.softmax(h1)
    # CategoricalCrossEntropy -> 1
    #h1 = F.categorical_cross_entropy(h1, y)
    return h1

#カメラスタート
while True:
    ret, frame = cap.read()
    cv2.imshow("camera", frame)
    k = cv2.waitKey(1)&0xff # キー入力を待つ
    
    if k == ord('r') or  k == ord('n') or k == ord('l') :
        if k ==114 :
            correct_label = "right"
        elif k == 110:
            correct_label = "none"
        elif k == 108:
            correct_label = "left"
        print (correct_label)
        # 「p」キーで画像を保存
        date = datetime.now().strftime("%Y%m%d_%H%M%S")
        #path = "./img/" + date + ".png"
        #path = new_dir_path +"./"+correct_label+"./"+ date +".png"
        path = new_dir_path +"./"+correct_label +".png"
        cv2.imwrite(path, frame)       

        
        image_gs = cv2.imread(path)
        
        # 顔認識の実行 顔だけ切り出して保存
#         face_list = cascade.detectMultiScale(image_gs,scaleFactor=1.2,minNeighbors=1,minSize=(130,130))
                
#         no = 1;
#         if len(face_list) == 0:
#             print("no face")
#             continue

        image_gs = cv2.cvtColor(image_gs,cv2.COLOR_BGR2GRAY)           
    
        dst = cv2.resize(frame, (int(frame.shape[1]/4), int(frame.shape[0]/4)))


        #path = new_dir_path +"./"+correct_label+"./"+ date +".png"
        path = new_dir_path +"./"+correct_label +"after.png"

        a = cv2.imwrite(path, dst)
        #plt.show(plt.imshow(np.asarray(Image.open(path))))
        print(no)
        no += 1

    #cv2.imwrite(path, frame) # ファイル保存

    #cv2.imshow(path, frame) # キャプチャした画像を表示


        f = pd.DataFrame(columns=["x:data","y:data"])
        xdata = path
        ydata = 0
        new_name = pd.Series([xdata,ydata],index=f.columns)
        f = f.append(new_name, ignore_index=True)
        f.to_csv('valu.csv',index=False,header = True )


        test_data = data_iterator_csv_dataset("valu.csv",1,shuffle=False,normalize=True) 
            
                 #ネットワークの構築
        nn.clear_parameters()
        #         # Prepare input variable
        x = nn.Variable((1,1,60,120))
        t = nn.Variable((1,1))
        #         # Build network for inference
        y = network(x, t)
        # #nn.load_parameters(r'C:\Users\jump1268\sony\nano\5actionCNN.files\20181212_170919\parameters.h5')
        #nn.load_parameters(r'C:\Users\lab\jupyternote\camera\img\cutface\NNC\nnca.files\20190418_171913\results.nnp')
        print("load model")
            # # print(arr)

        for i in range(test_data.size):
                x.d, t.d = test_data.next()
                y.forward()
                print(y.d[0])
                label = 2 #初期化
                percent = 1 #初期化
                


                if y.d[0][0]>y.d[0][1] and y.d[0][0]>y.d[0][2]:
                    print("left")
                    label = 0
                    result_label = "left"
                    percent = y.d[0][0]
                    #label = random.randint(0,2)
                if y.d[0][1]>y.d[0][0] and y.d[0][1]>y.d[0][2] :
                    print("none")
                    label = 1
                    result_label = "none"
                    percent = y.d[0][1]
                if y.d[0][2]>y.d[0][1] and y.d[0][2]>y.d[0][0] :
                    print("right")
                    label = 2
                    result_label = "right"
                    percent = y.d[0][2]     

                
                result = str(label)
              #  print (result)
                percent1 = percent*100           
                percent1 =str(percent1)
                print (percent1+"％")
                if result_label == correct_label :
                    print("正解")
                    new_correct_path =  "./realtime/correct"
                    os.makedirs(new_correct_path, exist_ok=True)
                    path = new_correct_path +"./"+correct_label+"./"+ date +".png"
                    a = cv2.imwrite(path, dst)                    
                else :
                    print("不正解")
                    new_uncorrect_path =  "./realtime/uncorrect"
                    os.makedirs(new_uncorrect_path, exist_ok=True)
                    path = new_uncorrect_path +"./"+result_label+"./"+correct_label+ date +".png"
                    a = cv2.imwrite(path, dst)
                    
                    real_path = new_uncorrect_path +"./"+"real_"+correct_label+"./"+correct_label+ date +".png"
                    a = cv2.imwrite(real_path, dst)
                    
    elif k == ord('q'):
        # 「q」キーが押されたら終了する
        break

# キャプチャをリリースして、ウィンドウをすべて閉じる
cap.release()
cv2.destroyAllWindows()
```
